```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
male = pd.read_excel('./体测分数_男生.xls')
female = pd.read_excel('./体测分数_女生.xls')
display(male, female)
```

```python
# 对男1000米跑、男引体进行等宽分箱操作，分成3份，并使用饼图绘制百分比

# 男1000米跑
male1000 = pd.cut(male['男1000米跑'], bins = 3)
male1000

dataCount = male1000.value_counts()

plt.figure(figsize=(9,9))



_ = plt.pie(dataCount,
            labels= dataCount.index,
            textprops={'family':'STSONG','fontsize':18}, # 设置字体样式
            autopct='%0.2f%%', # 显示百分比
#             explode = [0.1,0,0,0,0],# 突出某一部分
            shadow=True)

_ = plt.title(label='         1000米跑          ',family = 'STSONG', fontsize = 32,
              pad = 10, backgroundcolor='#c5b783',color='white',weight='bold')


# 男引体
maleYinti = pd.cut(male['男引体'], bins = 3)
maleYinti

dataCount = maleYinti.value_counts()

plt.figure(figsize=(9,9))



_ = plt.pie(dataCount,
            labels= dataCount.index,
            textprops={'family':'STSONG','fontsize':18}, # 设置字体样式
            autopct='%0.2f%%', # 显示百分比
#             explode = [0.1,0,0,0,0],# 突出某一部分
            shadow=True)

_ = plt.title(label='         男引体       ',family = 'STSONG', fontsize = 32,
              pad = 10,backgroundcolor='#c5b783',color='white',weight='bold')
```

```python
# 对女800米跑、女跳远进行直方图绘制统计各分数段人数，分成4份
#女800米跑
labels=['差','可','良', '优']
colors = ['#ff0000', '#ffa500', '#c5b783', '#3c7f99']
female800 = pd.qcut(female['女800米跑'], q = 4, labels=labels)

data = female800.value_counts()
data
plt.rcParams['font.family'] = 'STSONG'
plt.figure(figsize=(12,9))
bars  = plt.bar(x=data.index, height=data, tick_label = labels, color = colors,width = 0.35)

for i,bar in enumerate(bars):
    h = bar.get_height()
    w = bar.get_width()
    plt.text(x = bar.get_x() + w/2,
            y = h + 0.5,
            s = data[i],
            ha = 'center')


plt.title(label='      女800米跑       ', fontsize = 32,
          color = 'white',backgroundcolor = '#c5b783',weight = 'bold',pad = 30)

plt.grid(axis = 'y')
plt.box(False)

# 女跳远
labels=['差','可','良', '优']
colors = ['#ff0000', '#ffa500', '#c5b783', '#3c7f99']
female800 = pd.qcut(female['女跳远'], q = 4, labels=labels)

data = female800.value_counts()
data
plt.rcParams['font.family'] = 'STSONG'
plt.figure(figsize=(12,9))
bars  = plt.bar(x=data.index, height=data, tick_label = labels, color = colors,width = 0.35)

for i,bar in enumerate(bars):
    h = bar.get_height()
    w = bar.get_width()
    plt.text(x = bar.get_x() + w/2,
            y = h + 0.5,
            s = data[i],
            ha = 'center')


plt.title(label='      女跳远       ', fontsize = 32,
          color = 'white',backgroundcolor = '#c5b783',weight = 'bold',pad = 30)

plt.grid(axis = 'y')
plt.box(False)
```
