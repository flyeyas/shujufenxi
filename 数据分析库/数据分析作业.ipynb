```python
import numpy as np
```

```python
# 随机数生成六个班的考试成绩，3门考试：Python、数学、语文。每个班50人
scoreArr = np.random.randint(0,151, size = (6, 50, 3))
```

```python
# 将六个班的考试成绩进行合并得到score
score = np.concatenate(scoreArr)
score
```

```python
# 生成性别数组sex，水平叠加数组sex和score得到data
sexArr = np.random.randint(0,2, size = (300,1))
sexArr
data = np.concatenate([sexArr,score], axis = 1)
data
```

```python
# 分别计算男女生各科成绩统计指标：最小值、最大值、平均分、中位数、标准差
# 男女分组
male = np.empty(shape = (0,3), dtype = int)
female = np.empty(shape = (0,3), dtype = int)
display(male, female)

for row in range(0,300):
    if data[row,0] == 0:
        female = np.append(female,[data[row,1:]],axis = 0)
    elif data[row,0] == 1:
        male = np.append(male,[data[row,1:]],axis = 0)
# display(male, female)

# 女生各科成绩统计指标
scoreMin = female.min(axis = 0)
scoreMax = female.max(axis = 0)
scoreMean = female.mean(axis = 0)
scoreMedian = np.median(female,axis = 0)
scoreStd = female.std(axis = 0)
print("女生各科成绩统计")
print("\tPython\t数学\t语文")
print("最小值\t",scoreMin[0],"\t",scoreMin[1],"\t",scoreMin[2])
print("最大值\t",scoreMax[0],"\t",scoreMax[1],"\t",scoreMax[2])
print("平均值\t",round(scoreMean[0],2),"\t",round(scoreMean[1],2),"\t",round(scoreMean[2],2))
print("中位数\t",scoreMedian[0],"\t",scoreMedian[1],"\t",scoreMedian[2])
print("标准差\t",round(scoreStd[0],2),"\t",round(scoreStd[1],2),"\t",round(scoreStd[2],2))

# 男生生各科成绩统计指标
scoreMin = male.min(axis = 0)
scoreMax = male.max(axis = 0)
scoreMean = male.mean(axis = 0)
scoreMedian = np.median(male,axis = 0)
scoreStd = male.std(axis = 0)
print("男生各科成绩统计")
print("\tPython\t数学\t语文")
print("最小值\t",scoreMin[0],"\t",scoreMin[1],"\t",scoreMin[2])
print("最大值\t",scoreMax[0],"\t",scoreMax[1],"\t",scoreMax[2])
print("平均值\t",round(scoreMean[0],2),"\t",round(scoreMean[1],2),"\t",round(scoreMean[2],2))
print("中位数\t",scoreMedian[0],"\t",scoreMedian[1],"\t",scoreMedian[2])
print("标准差\t",round(scoreStd[0],2),"\t",round(scoreStd[1],2),"\t",round(scoreStd[2],2))
```
